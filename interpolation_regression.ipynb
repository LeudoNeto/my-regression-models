# Interpolation Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
x = dataset.iloc[:, -2].values
y = dataset.iloc[:, -1].values
```

## Training the Interpolation model on the whole dataset

```python
from main import InterpolationRegression
regressor = InterpolationRegression()
regressor.fit(x, y)
```

## Visualising the Interpolation Regression results

```python
plt.scatter(x,y, color='red')
plt.plot(x, regressor.predict(x))
plt.title('Truth of Bluff')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Interpolation Regression results (for higher resolution and smoother curve)

```python
x_grid = np.arange(min(x), max(x), 0.1)
x_grid = x_grid.reshape((len(x_grid), 1))
plt.scatter(x, y, color = 'red')
plt.plot(x_grid, regressor.predict(x_grid), color = 'blue')
plt.title('Truth or Bluff ')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

## Predicting a new result with Interpolation Regression

```python
print(regressor.predict([[7.5]]))
```

## Comparing to Linear Regression Model

### Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

### Training and ranking both models

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

int_regressor = InterpolationRegression()
int_regressor.fit(x_train, y_train)

print(r2_score(y_test, int_regressor.predict(x_test)))

x_train, y_train = x_train.reshape(-1,1), y_train.reshape(-1,1)
x_test, y_test = x_test.reshape(-1,1), y_test.reshape(-1,1)

lin_regressor = LinearRegression()
lin_regressor.fit(x_train, y_train)

print(r2_score(y_test, lin_regressor.predict(x_test)))
```
